```python
import cv2

# Load the pre-trained Haar Cascade classifier for full body detection
classifier = cv2.CascadeClassifier("haarcascade_fullbody.xml")

# Initialize variables to keep track of the number of people detected
people_count = 0

# Initialize list to keep track of the positions of people in the video
people_positions = []

# Read the input video
cap = cv2.VideoCapture("sample.mp4")

# Create a new window for the video
cv2.namedWindow("Frame", cv2.WINDOW_NORMAL)

# Move the window to the second screen
cv2.moveWindow("Frame", 1920, 0)

while True:
    # Read a frame from the video
    ret, frame = cap.read()
    if not ret:
        break

    # Convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Pass the frame through the classifier
    rects = classifier.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=3, minSize=(30, 30))

    # Loop through the detections
    for (x, y, w, h) in rects:
        # Create a bounding box around the detected person
        x1, y1, x2, y2 = x, y, x+w, y+h
        center = (int((x1+x2)/2), int((y1+y2)/2))
        
        # check if the center of the bounding box is already in people_positions list
        if center in people_positions:
            continue
        else:
            people_count += 1
            people_positions.append(center)
            cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)
        
    # Show the frame
    cv2.imshow("Frame", frame)
    key = cv2.waitKey(1)
    if key == 27:
        break

    # Print the number of people
    print("Number of people detected:", people_count)

# Release the video capture and close the window
cap.release()
cv2.destroyAllWindows()
```
